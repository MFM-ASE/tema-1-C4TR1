```python
import yfinance as yf
import numpy as np
import pandas as pd

#define the ticker symbol
tickerSymbol = 'CL=F'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
df = tickerData.history(period='1d', start='2022-1-1', end='2022-10-16')

#see your data
df
```

```python
#define the ticker symbol
tickerSymbol2 = 'GC=F'

#get data on this ticker
tickerData2 = yf.Ticker(tickerSymbol2)

#get the historical prices for this ticker
df2 = tickerData2.history(period='1d', start='2022-1-1', end='2022-10-16')

#see your data
df2

df2.rename(columns = {'Open':'Open2', 'High':'High2', 'Low':'Low2', 'Close':'Close2', 'Volume':'Volume2', 'Dividends':'Dividends2', 'Stock Splits':'Stock Splits2'}, inplace = True)
df2
```

```python
dataframe = pd.concat([df,df2], axis=1)
dataframe
```

```python
close = dataframe[["Close", "Close2"]]
print(close)

simple_return = close.pct_change()
print(simple_return)

log_return = np.log(1+simple_return)
print(log_return)
```
